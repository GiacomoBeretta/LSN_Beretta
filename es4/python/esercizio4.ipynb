Ho fatto partire la simulazione con la temperatura in unit√† di Lennard Jones a T=1,544, dalla configurazione a reticolo scritta nel file config.fcc, in questo modo dopo 100 blocchi da 1000 passi, ho ottenuto la temperatura desiderata di 0,8. A questo punto per portare il sistema all'equilibrio ho realizzato 10 milioni di passi e notando dal grafico della temperatura che dopo circa 6 milioni di passi la temperatura si stabilizzava attorno al valore desiderato ho fatto ripartire la simulazione utilizzando come configurazione iniziale quella finale del ciclo precedente e come temperatura T=0.8. Per quest'ultima simulazione ho usato 100 blocchi da 1000 passi

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

MDMC, restart, T, Npart, rho, rcut, delta, nblk, nstep, nstart = np.loadtxt("./input.solid")

E_particle = np.loadtxt("./etot_solid.dat")
U_particle = np.loadtxt("./epot_solid.dat")
K_particle = np.loadtxt("./ekin_solid.dat")
Temp = np.loadtxt("./temp_solid.dat")
Pres = np.loadtxt("./pres_solid.dat")

x = np.arange(nblk+1)
x = x*nstep

fig1, axs = plt.subplots(5, 1, tight_layout = False, figsize=(14, 12))

axs[0].errorbar(x[1:], E_particle[:,2], yerr = E_particle[:,3])
axs[1].errorbar(x[1:], U_particle[:,2], yerr = U_particle[:,3])
axs[2].errorbar(x[1:], K_particle[:,2], yerr = K_particle[:,3])
axs[3].errorbar(x[1:], Temp[:,2], yerr = Temp[:,3])
axs[4].errorbar(x[1:], Pres[:,2], yerr = Pres[:,3])

axs[0].set_ylabel(r'$E_{tot}/\epsilon}$')
axs[1].set_ylabel(r'$U_{tot}/\epsilon}$')
axs[2].set_ylabel(r'$K_{tot}/\epsilon}$')
axs[3].set_ylabel(r'$TK_B/\epsilon}$')
axs[4].set_ylabel(r'$P\sigma^3/\epsilon}$')
    
axs[0].set_title("total energy per particle")
axs[1].set_title("potential energy per particle")
axs[2].set_title("kinetic energy per particle")
axs[3].set_title("temperature")
axs[4].set_title("pressure")

for i in range(5):
    axs[i].grid(True)
    axs[i].set_xlabel('number of steps')
    
plt.suptitle("Molecular Dynamics simulation of 108 particles of Argon in a solid state\nin periodic boundary conditions with Lennard Jones potential, using " + str(nblk) + " blocks of " + str(nstep) + " steps", size='xx-large')
fig1.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0)
plt.show()
```
