```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math

data = np.array
data = np.loadtxt("./random_mean_dev.txt")
parameters = np.loadtxt("./test_parameters.txt")
N = parameters[0]#numero di random totali
M = parameters[1]#numero di blocchi
L = N/M
print("N=", N)
print("M=", M)
print("L=", L)

x = np.arange(M+1)
x = x*L

plt.errorbar(x[1:], data[:,0] - 0.5, yerr = data[:,1])
plt.xlabel('#throws (N)')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.title("mean values for a uniform distribution in (0,1) with their standard deviation of the mean as the number N of random numbers grows")
plt.show()
```

3) Test del $\chi^2$ per verificare una distribuzione (pseudo) uniforme in (0,1) dei numeri casuali generati. Il calcolo  stato fatto per blocchi di $10^4$ numeri random, dopo aver diviso l'intervallo (0,1) in 100 parti. Dato che l'intervallo (0,1)  stato diviso in 100 parti il numero di classi  100. Il numero $\nu$ di gradi di libert per la distribuzione del $\chi^2$  dato dalla formula $\nu = n_{classi} - c - 1 $ dove $c$  il numero di parametri calcolati dal campione che in questo caso  0 dato che i valori attesi sono stati calcolati a priori, e -1 perch  sempre presente la condizione $ \sum_{j}^{n_{classi}}{x_j} = N$. Nel nostro caso quindi $\nu = 100 - 1 = 99$.

```python
from scipy.stats import chi2
data_chi = np.array
data_chi = np.loadtxt("./random_chi_quadro.txt")

N_bins_chi = data_chi[0]
N_chi = data_chi[1]

v = N_bins_chi-1 #degrees of freedom
N_bins_histo = 100

chi_max = max(data_chi[2:])
chi_min = min(data_chi[2:])

fig, ax = plt.subplots(1, 1)

x = np.linspace(chi2.ppf(0.01, v), chi2.ppf(0.99, v), 100)

bin_width = (chi_max - chi_min) / N_bins_histo
ax.plot(x, chi2.pdf(x, v) * bin_width * N_chi, 'r-', lw=2)

plt.hist(data_chi[2:], bins=N_bins_histo)
plt.xlabel('$\chi^2 values$')
plt.ylabel('counts')

plt.title(f"distribution of $\chi^2$ for 10000 random numbers from a uniform distribution in (0,1)")
plt.show()
```

```python
%%time
#caricamento dati
data_dice = np.array
data_dice = np.loadtxt("./Standard_Dice.txt")

data_exp = np.array
data_exp = np.loadtxt("./Exponential.txt")

data_cauchy = np.array
data_cauchy = np.loadtxt("./Cauchy_Lorentz.txt")

M = 10000 #numero iterazioni
#creo 4 array per ciascuno che conterranno i dati per i 4 valori di N=1,2,10,100
dice_1 = data_dice[0:M]
dice_2 = data_dice[M:2*M]
dice_10 = data_dice[2*M:3*M]
dice_100 = data_dice[3*M:4*M]

exp_1 = data_exp[0:M]
exp_2 = data_exp[M:2*M]
exp_10 = data_exp[2*M:3*M]
exp_100 = data_exp[3*M:4*M]

cauchy_1 = data_cauchy[0:M]
cauchy_2 = data_cauchy[M:2*M]
cauchy_10 = data_cauchy[2*M:3*M]
cauchy_100 = data_cauchy[3*M:4*M]

#creo gli istogrammi
n_bins = [6, 11, 51, 501]

#standard dice
N = (1, 2, 10, 100)

fig1, axs1 = plt.subplots(2, 2, tight_layout = True, figsize=(16,16))

axs1[0, 0].hist(dice_1,   bins=n_bins[0])
axs1[0, 1].hist(dice_2,   bins=n_bins[1])
axs1[1, 0].hist(dice_10,  bins=n_bins[2])
axs1[1, 1].hist(dice_100, bins=n_bins[3])

i = 0
for ax in axs1.flat:
    ax.set(xlabel='values of S_N', ylabel='number of counts')
    ax.set_title(f'{M} mean values for a standard dice with {N[i]} extractions, distributed in {n_bins[i]} bins')
    i += 1

    
# exponential dice
n_bins = [1000, 1000, 1000, 1000]
fig2, axs2 = plt.subplots(2, 2, tight_layout = True, figsize=(16,16))

axs2[0, 0].hist(exp_1,   bins=n_bins[0])
axs2[0, 1].hist(exp_2,   bins=n_bins[1])
axs2[1, 0].hist(exp_10,  bins=n_bins[2])
axs2[1, 1].hist(exp_100, bins=n_bins[3])

i = 0
for ax in axs2.flat:
    ax.set(xlabel='values of S_N', ylabel='number of counts')
    ax.set_title(f'{M} mean values for an exponential dice with {N[i]} extractions, distributed in {n_bins[i]} bins')
    i += 1
    
#Cauchy-Lorentz dice
n_bins = [1000, 1000, 1000, 1000]
fig3, axs3 = plt.subplots(2, 2, tight_layout = True, figsize=(16,16))

axs3[0, 0].hist(cauchy_1,   bins=n_bins[0], range=(-10, 10))
axs3[0, 1].hist(cauchy_2,   bins=n_bins[1], range=(-10, 10))
axs3[1, 0].hist(cauchy_10,  bins=n_bins[2], range=(-10, 10))
axs3[1, 1].hist(cauchy_100, bins=n_bins[3], range=(-10, 10))

i = 0
for ax in axs3.flat:
    ax.set(xlabel='values of S_N', ylabel='number of counts')
    ax.set_title(f'{M} mean values for a Cauchy-Lorentz dice with {N[i]} extractions, distributed in {n_bins[i]} bins ')
    #ax.set_xscale('symlog')
    i += 1

plt.show()
```

# Esercizio 1.3

Per l'esercizio del calcolo del $\pi$ ho scelto come distanza tra gli assi orizzontali d = 2 e lunghezza dell'ago L_needle = 1.
Per "lanciare" a caso l'ago ho usato la seguente procedura: ho estratto tramite una distribuzione uniforme tra (0, d) le coordinate $x_0, y_0$ di un estremo dell'ago, poi ho fatto la stessa cosa per un altro punto a caso $x_1, y_1$, in questo modo la congiungente dei due punti d una direzione casuale, per avere infine le coordinate del secondo estremo dell'ago $x_2, y_2$ ho fatto in modo che la distanza tra i due estremi $x_0, y_0$ e $x_2, y_2$ fosse uguale a L_needle. Trovata $y_2$ se $y_2 > d$ oppure $y_2 < 0$ vuol dire che l'ago interseca una linea orizzontale. Ho ripetuto la procedura per N_throws = 1000 volte, calcolando cos $\pi$ con la formula $$\pi = \frac{2LN_{thr}}{N_{hit}d}$$.

```python
data_pi = np.array
data_pi = np.loadtxt("./pi_greco.txt")

parameters_pi = np.loadtxt("./pi_greco_parameters.txt")

N = parameters_pi[0] #numero di realizzazioni
M = parameters_pi[1] #numero di blocchi
L_needle = parameters_pi[2]
d = parameters_pi[3]
N_throws = parameters_pi[3]

L = N/M

x = np.arange(M+1)
x = x*L

plt.errorbar(x[1:], data_pi[:, 0], yerr = data_pi[:, 1])
plt.xlabel('#realizations (N)')
plt.ylabel('x')
plt.grid(True)
plt.title("mean values for the calculation of pi (with the method of Buffon needle) with their standard deviation of the mean as the number N of calculated means grows (method without using $\pi$)")
plt.axhline(math.pi, c='red', lw=1, ls='solid')
plt.show()
```
